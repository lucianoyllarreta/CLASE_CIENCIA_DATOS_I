### Paso 1: Filtrar los últimos 20 años y asignar la región
Creamos un nuevo DataFrame con datos desde 2004 y aplicamos las listas de regiones definidas anteriormente.

```
# 1. Filtrar por año
df_20_years = df[df['year'] >= 2004].copy()

# 2. Función simple para asignar región
def definir_region(pais):
    if pais in america_del_sur: return 'América del Sur'
    if pais in america_central_y_caribe: return 'Centroamérica y Caribe'
    if pais in america_del_norte: return 'América del Norte'
    if pais in europa: return 'Europa'
    if pais in africa: return 'África'
    if pais in asia: return 'Asia'
    if pais in oceania: return 'Oceanía'


df_20_years['region'] = df_20_years['country'].apply(definir_region)
```

```
# 1. Filtrar por año
df_20_years2 = df[df['year'] >= 2004].copy()

# 2. Función simple para asignar región
def definir_region(pais):
    if pais in mercosur: return 'Mercosur'
    if pais in brics: return 'Brics'
    if pais in medio_oriente: return 'Medio Oriente'
    if pais in union_europea: return 'Union Europea'
    if pais in ocde: return 'OCDE'
    if pais in europa_no_ue: return 'Europa no UE'
df_20_years2['region'] = df_20_years2['country'].apply(definir_region)
```

```
# Agrupamos y calculamos la media
resumen_regiones = df_20_years.groupby('region')[['gdp', 'population']].mean()

display(resumen_regiones)
```

```
resumen_bloqueseconomicos = df_20_years2.groupby('region')[['gdp', 'population']].mean()

display(resumen_bloqueseconomicos)
```

```
# Cálculo final
resumen_regiones['pbi_per_capita_promedio'] = resumen_regiones['gdp'] / resumen_regiones['population']

# Mostramos el resultado ordenado
display(resumen_regiones[['pbi_per_capita_promedio']].sort_values(by='pbi_per_capita_promedio', ascending=False))
```

```

resumen_bloqueseconomicos['pbi_per_capita_promedio'] = resumen_bloqueseconomicos['gdp'] / resumen_bloqueseconomicos['population']


display(resumen_bloqueseconomicos[['pbi_per_capita_promedio']].sort_values(by='pbi_per_capita_promedio', ascending=False))
```
